Undersampling 

Build a sample dataset containing normal and fraudulent transaction

Fraudulent transaction --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating 2 datasets ie, legit_sample & fraud

```python
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
 new_data.groupby('Class').mean()
```

splitting data into features and target

```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
print(X)
```

split the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state="4. CNN CIFAR 10 dataset.ipynb"2) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training 

## Logistics regression

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy score :", training_accuracy)
```

```python
# accuracy on test data 
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy score:", testing_accuracy)
```
